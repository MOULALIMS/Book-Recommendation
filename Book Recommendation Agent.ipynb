```python
ratings =pd.read_csv('BX-Book-Ratings.csv', sep=";",on_bad_lines='skip',encoding='latin-1')
```

```python
ratings.head()
```

```python
ratings.rename(columns={
    "User-ID" : "user_id",
    "Book-Rating" : "rating"},inplace=True)
```

```python
ratings['user_id'].value_counts()
```

```python
print(ratings.shape)
```

```python
ratings['user_id'].unique().shape
```

```python
x =ratings['user_id'].value_counts() > 200
x
```

```python
x[x].shape
```

```python
y= x[x].index
y
```

```python
ratings = ratings[ratings['user_id'].isin(y)]
ratings.head()
```

```python
ratings.shape
```

```python
books.head()
```

```python
ratings_with_books = ratings.merge(books, on="ISBN")
ratings_with_books.head()
```

```python
ratings_with_books.shape
```

```python
num_rating = ratings_with_books.groupby('title')['rating'].count().reset_index()
num_rating
```

```python
num_rating.rename(columns={
    "rating" : "num_of_ratings"
},inplace=True)
```

```python
num_rating.head()
```

```python
final_rating= ratings_with_books.merge(num_rating,on='title')
final_rating
```

```python
final_rating = final_rating[final_rating['num_of_ratings']>=50]
```

```python
final_rating.sample(10)
```

```python
final_rating.shape
```

```python
final_rating.drop_duplicates(['user_id','title'],inplace=True)
final_rating.shape
```

```python
book_pivot = final_rating.pivot_table(columns='title', index='user_id', values='rating')
book_pivot
```

```python
book_pivot.shape
```

```python
book_pivot.fillna(0,inplace=True)
book_pivot
```

```python
#CSR matrix is used to avoid the zeroes in pivot table for faster consideration 
from scipy.sparse import csr_matrix
```

```python
book_sparse = csr_matrix(book_pivot)
book_sparse
```

```python
from sklearn.neighbors import NearestNeighbors
model = NearestNeighbors(algorithm='brute')
```

```python
model.fit(book_sparse)
```

```python


distance , suggestion = model.kneighbors(book_pivot.iloc[237,:].values.reshape(1,-1), n_neighbors=6)
print(distance)
print(suggestion)

    
```

```python

for i in range(len(suggestion)):
    print(book_pivot.columns[suggestion[i]])
```

```python
def recommend_book(book_name):
    book_id = np.where(book_pivot.columns == book_name)[0][0]
    distance, suggestion = model.kneighbors(book_pivot.iloc[book_id, :].values.reshape(1, -1), n_neighbors=6)


    print("Number of columns in book_pivot:", len(book_pivot.columns))

    for i in range(len(suggestion)):
        books = book_pivot.columns[suggestion[i]]
        for j in books:
            print(j)
```

```python
book_name = "1984"
recommend_book(book_name)
```
